# Working in Google Colab

Every notebook in this program runs in Google Colab. Before you open your first
research question, this short tutorial gets you comfortable with the one place
you'll do all of your work. You don't need to know any Python yet — you'll learn
that in the next tutorial. Right now the only goal is to feel at home in Colab:
how to run a cell, where your work is saved, and how to get the data and tools
this program depends on.

Run each cell as you reach it (we'll show you how in a moment) and read the notes
in between. By the end you'll have set up your environment for real and seen actual
plant data on your screen.

### 1) What Colab is, and why we use it

The research you'll do this summer means running code — loading data, training
models, making figures. That code needs a computer with the right software
installed and, for some of the image work, a powerful graphics chip (a GPU).
Setting all of that up on your own laptop is fiddly, and not everyone has a
machine that can do it.

Google Colab solves this. It's a free service that runs notebooks like this one
in your web browser, on Google's computers instead of your own. There's nothing
to install. You open a notebook, it connects to a machine in the cloud, and you
run your code there. Every student in this program works the same way, on the
same setup, which also means that when you ask for help, the person helping you
is looking at exactly the environment you're in.

A Colab notebook is a list of **cells** stacked top to bottom. Some cells are
text (like this one). Others hold code you can run. The next section shows you
how running works.

### 2) Cells and run order

There are two kinds of cells:

- **Text cells** (also called markdown cells) hold explanations like this one.
  They don't run anything.
- **Code cells** hold Python. When you run a code cell, Colab sends the code to
  the machine in the cloud, runs it, and shows you the result underneath.

To run a code cell, click it and press **Shift + Enter**, or click the round
play button on its left edge. The cell just below this one is a code cell — run
it now.

**Order matters.** Colab runs cells in the order you run them, not the order
they appear on the page. A cell that uses something defined in an earlier cell
will fail if you skipped that earlier cell. The safe habit, especially after a
break or a restart, is **Runtime → Run all**, which runs every cell from the top
in order. We'll come back to that in the last section.

```
# Your first code cell. Run it with Shift+Enter.
print("This ran on a computer in the cloud, not on your laptop.")
2 + 2
```

You should see the printed line, and then `4` on its own — Colab shows the value
of the last line in a cell automatically. That's the whole loop: write code, run
the cell, read what comes back.

### 4) Where your work lives

Here's the most important thing to understand about Colab, because getting it
wrong is the most common way students lose a day of work.

The cloud machine you're using is **temporary**. When your session ends — you
close the tab for too long, or Colab disconnects you after a period of inactivity
— that machine is wiped clean. Anything saved only to the machine itself is gone.

That's why `iri.setup()` connected your Google Drive. Your Drive is permanent and
belongs to you, so the program saves your results there instead of on the
throwaway machine. You don't have to manage any of this by hand. When a notebook
needs to save something, it asks the library for the right folder:

```
# Ask the library where this question's results should go.
# The string is just a label for the question; "r1_q1" is the first one.
folder = iri.output_dir("r1_q1")
print(folder)
```

The path that prints is a folder on your Google Drive, created for you if it
didn't already exist. A notebook writes its results there, and a later notebook
in the same question reads them back from the same place. Because it's on Drive,
those results survive a disconnect — you can come back tomorrow, reconnect, and
your work is still there.

You won't normally type these paths yourself. You just call `iri.output_dir(...)`
with the question's label and let it hand you the right folder. (We only printed
it here so you can see what it points to — we didn't save anything.)

### 5) Getting the data

In your research questions you won't download data files by hand. The library
has loader functions that fetch the right dataset, check it arrived intact, and
hand it to you ready to use. To make this concrete, let's load a small sample of
the PlantVillage dataset — a collection of leaf photographs you'll work with in
Rationale 2 — and actually look at a leaf.

This downloads about 3 MB the first time and is cached on your Drive afterward,
so a second run is instant.

```
data = iri.load_plantvillage_orientation()
data.keys()
```

The loader handed back a small bundle (a Python dictionary) with a few pieces
inside it. One of them, `sample_paths`, is a table listing sample images and
what each one is. Let's open the very first image and display it.

```
from PIL import Image
import matplotlib.pyplot as plt

first = data['sample_paths'].iloc[0]
image = Image.open(first['path'])

plt.imshow(image)
plt.title(f"{first['class']}")
plt.axis('off')
plt.show()
```

That's a real leaf from the dataset, loaded with one function call and no manual
downloading. This is the shape of everything you'll do: a loader brings the data
in, and you work with what it returns. The Rationale 1 and Rationale 2
orientation notebooks go much deeper into what these datasets contain and why —
this was just to show you that real data is one line away.

### 6) When you need a GPU

A GPU is a chip built for the heavy number-crunching that training image models
involves. You don't need one for everything — the gene-expression work in
Rationale 1 runs fine without it — but the image work in Rationale 2 trains much,
much faster on a GPU, and some notebooks expect one.

Colab gives you a GPU for free, but you have to ask for it. It isn't on by
default. To switch it on:

**Runtime → Change runtime type → Hardware accelerator → GPU → Save.**

Changing the runtime restarts your session, so do it *before* you run the install
and setup cells (or run them again afterward).

You can always check what you currently have:

```
print(f"GPU available right now: {iri.has_gpu()}")
```

If a notebook truly requires a GPU, its setup line will say so by using
`iri.setup(gpu_required=True)`. When you run that on a machine without a GPU, it
stops immediately with a clear message telling you to switch the runtime, rather
than letting you wait through a long job that was never going to finish. That's a
feature: it fails fast and tells you exactly what to fix.

(We won't run `gpu_required=True` here, since this tutorial is meant to work on
the ordinary free runtime.)

### 7) Sessions, restarts, and disconnects

A Colab **session** is your live connection to a cloud machine. A few things will
end it, and it helps to know what each one costs you:

- **Disconnecting** (closing the tab too long, or sitting idle). The machine is
  wiped. Anything you saved with `iri.output_dir(...)` is safe on your Drive;
  anything saved only to the machine is lost. To get back to work: reopen the
  notebook, reconnect, run the install and setup cells, then **Runtime → Run all**.
- **Restarting the runtime** (Runtime → Restart session, or the "RESTART SESSION"
  banner after an install). This clears everything Python is holding in memory
  but keeps your installed library and files for the moment. After a restart you
  generally want **Run all** so every variable is rebuilt in order.
- **Factory reset** (Runtime → Disconnect and delete runtime). A completely fresh
  machine. Same recovery as a disconnect: install, setup, run all.

The single habit that saves you from almost all session trouble: **when in doubt,
Runtime → Run all from the top.** Because your data loads from a cache and your
results live on Drive, re-running is cheap and reliable.

On the free tier, sessions also have time limits and can disconnect if you step
away. None of this loses your real work, as long as your results went to
`iri.output_dir(...)` — which, in this program, they always do.

### 8) Getting unstuck

Things will break sometimes. That's normal, and it's usually quick to fix. A
short checklist:

- **A cell failed and mentions something "not defined."** You probably ran cells
  out of order or after a restart. Do **Runtime → Run all** and try again.
- **A cell about the `iri` library failed.** Make sure you ran the install cell
  and the `import irilab2026 as iri` cell in this session.
- **Read the error from the bottom up.** The last line names what went wrong in
  plain-ish terms; the lines above it are the trail of where. You don't need to
  understand all of it — the bottom line is usually enough to know what to fix.
- **A long job stalled or the machine feels stuck.** Restart the runtime
  (Runtime → Restart session) and run all from the top.

When none of that works, copy the full error text and bring it to your mentor or
the program channel. Because everyone is on the same Colab setup, a shared error
is usually a shared, quick fix.
