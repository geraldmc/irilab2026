### 1) Variables and types

A **variable** is a name that holds a value. You create one by writing a name, an
equals sign, and the value. The `print` function shows you a value, and the `type`
function tells you what kind of value it is. Run the cell below.

```python
x = 2
print(x)
print(type(x))
```

You made a variable `x`, set it to `2`, printed it, and printed its type — `int`,
short for integer (a whole number). Numbers aren't the only kind of value. Three
other everyday types are text, decimals, and true/false:

```python
leaf = "healthy"   # a string (str): text, written in quotes
reading = 6.2      # a float: a number with a decimal point
is_stressed = True # a boolean (bool): either True or False

print(leaf)
print(reading)
print(is_stressed)
```

The `#` symbol starts a **comment** — a note for human readers that Python ignores.
Use comments to remind yourself what a line is for.

So the basic types are: `str` (text), `int` (whole numbers), `float` (decimals),
and `bool` (`True`/`False`). Variables are useful because of the things you can do
to them — that's next.

#### Operators

Python has **operators** for working with values: `+` and `-` add and subtract,
`*` and `/` multiply and divide, `**` raises to a power, and `==` checks whether
two things are equal. As an example, here's the formula for converting a
temperature from Fahrenheit to Celsius, which matters for plant experiments run in
labs that report in different units:

```python
fahrenheit = 96
celsius = 5 * (fahrenheit - 32) / 9
print(f"{fahrenheit} °F is {celsius:.1f} °C")
```

That output used an **f-string** — a string with an `f` in front, where anything
inside `{ }` is filled in with a value. The `:.1f` part rounded the number to one
decimal place. f-strings are how you'll build readable output throughout your work;
you'll see them constantly.

### 2) Lists

Organizing data well is half of doing science with a computer. The simplest way to
hold several things together is a **list**, written with square brackets. Here's a
list of the eight stress conditions you'll study in Rationale 1 — the different
ways plants were stressed in the experiment:

```python
stresses = ["cold", "osmotic", "salt", "drought",
            "genotoxic", "uv_b", "wounding", "heat"]
print(stresses)
print(f"There are {len(stresses)} stress conditions.")
```

`len` gave you the number of items. To reach a single item, you use its position in
square brackets — but here's the part that trips up newcomers: Python counts from
**zero**. The first item is at position `0`, the second at position `1`, and so on.

```python
print(stresses[0])  # the first item
print(stresses[1])  # the second item
```

Because counting starts at zero, the last item of an eight-item list is at position
`7`, not `8`. Asking for a position that doesn't exist is an error. The cell below
deliberately asks for position `8` to show you what that looks like — we've wrapped
it so it reports the error instead of stopping the notebook:

```python
try:
    print(stresses[8])
except IndexError as error:
    print(f"IndexError: {error}")
    print("There's no position 8 — the valid positions are 0 through 7.")
```

An `IndexError` means you reached past the end of the list. When you see one in your
own work, it almost always means an off-by-one mistake or a list shorter than you
expected. (Out in the wild this error stops a program; we caught it here on purpose
so the rest of the notebook keeps running.)

### 3) Dictionaries

A list is *ordered* — you reach items by their position. A **dictionary** is
different: you reach items by a **key** you choose, usually a word. That makes
dictionaries perfect as lookup tables. Here's one that maps each of a few stress
names to a plain-English description:

```python
stress_descriptions = {
    "cold": "exposure to low temperature",
    "drought": "water removed quickly — fast desiccation, not slow field drought",
    "salt": "too much salt in the soil",
}

print(stress_descriptions["drought"])
```

You looked up the value for the key `"drought"`. (Notice the description there — in
this dataset "drought" is fast drying in the lab, not the slow drought of a field.
That's the kind of detail the orientation notebook will return to.)

Asking for a key that isn't in the dictionary is an error, much like a bad list
position. Again we'll catch it so the notebook keeps running:

```python
try:
    print(stress_descriptions["heat"])
except KeyError as error:
    print(f"KeyError: {error}")
    print("We never added a 'heat' description to this dictionary.")
```

A `KeyError` means the key you asked for isn't there. Lists and dictionaries are the
two containers you'll reach for most — a list when order and position matter, a
dictionary when you want to look things up by name. Hold on to both; the final
section shows how the program's data is, in a sense, both at once.

### 4) Making decisions and repeating work

Real analysis needs to make choices ("if the value is high, call it stressed") and
repeat steps ("do this for every stress condition"). Those are **control flow**:
`if` statements for choices, `for` loops for repetition.

An `if` statement runs a block of code only when its condition is `True`. You can
add `elif` ("else if") for more conditions, and `else` for everything left over:

```python
expression_level = 7.4

if expression_level > 8:
    print("Strongly expressed")
elif expression_level > 5:
    print("Moderately expressed")
else:
    print("Low or no expression")
```

Try changing `expression_level` and re-running to see the branch change.

A `for` loop walks through a collection one item at a time. Remember the `stresses`
list from earlier? You can visit each stress in turn:

```python
for stress in stresses:
    print(f"Studying the {stress} condition")
```

You can also loop a fixed number of times with `range`. `range(3)` gives you the
positions `0, 1, 2`:

```python
for i in range(3):
    print(f"Pass number {i}")
```

A handy trick in Colab: put a `?` after any function to see its built-in help. Run
the cell below to read what `range` does, then close the help panel.

```python
range?
```

### 5) Importing tools, and a word on reproducibility

Python comes with a large **standard library** — collections of ready-made tools you
load with `import`. You don't reinvent things like random-number generation; you
import them. Here's the `random` module producing a number between 0 and 1:

```python
import random

print(random.random())
```

Run that cell twice — you get a different number each time. That's a problem for
science: if your analysis uses randomness, no one (including you) can reproduce your
exact result later. The fix is a **seed**: a fixed starting point that makes the
"random" sequence come out the same every run.

This is why `iri.setup()` seeds randomness for you at the start of every notebook.
You don't have to manage it by hand — but now you know what it's doing and why it
matters. Reproducibility is something this program takes seriously, and seeding is
the first piece of it.

### 6) Functions

When you find yourself writing the same steps more than once, wrap them in a
**function** and call it by name. You define a function with `def`, list its inputs
in parentheses, and hand back a result with `return`:

```python
def fahrenheit_to_celsius(fahrenheit):
    return 5 * (fahrenheit - 32) / 9

print(fahrenheit_to_celsius(96))
print(fahrenheit_to_celsius(32))
```

Inputs can have **default values** — used when you don't supply your own. Here, a
function labels an expression reading, with an optional threshold that defaults to
`5`:

```python
def label_expression(value, threshold=5):
    if value > threshold:
        return "expressed"
    else:
        return "low"

print(label_expression(7.4))            # uses the default threshold of 5
print(label_expression(7.4, threshold=8))  # supplies a different threshold
```

`return` is what sends a value back out of the function so you can use it. A function
that just prints, with no `return`, hands back nothing. As your analysis grows,
moving repeated steps into well-named functions keeps it readable and keeps mistakes
in one place instead of scattered everywhere.
